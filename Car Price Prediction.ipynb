# Car Price Prediction

```python
# Workflow of Machine Learning Model
'''
0) Import Important dependencies
1) Setting up the data
2) Creating the Model
3) Training the Model
4) Testing the Model
5) Prediction
'''
```

## 1. Import libraries

```python
# Importing Important Dependencies
import numpy as np
import pandas as pd
```

```python
# Importing LinearRegression Model(to predict continuous values)
from sklearn.linear_model import LinearRegression
```

```python
# Importing train test split function
from sklearn.model_selection import train_test_split
```

```python
# Importing Accuracy score
from sklearn import metrics
```

## 2. Load Dataset

```python
# Import dataset
df = pd.read_csv('cardata.csv')
df
```

## 3. Data preprocessing

```python
# Info of the dataset
df.info()
```

```python
# Checking null values
df.isnull().sum()
```

```python
# Basic analysis
# Find the unique values under Fuel_Type, Seller_Type, Transmission
print(pd.unique(df['Fuel_Type']))
print(pd.unique(df['Seller_Type']))
print(pd.unique(df['Transmission']))
```

```python
# Feature Encoding --> converting non-numerical categorical data 
# (e.g., labels, names, types) into a numerical format
df.replace({'Fuel_Type': {'Petrol': 0, 'Diesel': 1, 'CNG': 2}}, inplace = True)
df.replace({'Seller_Type': {'Dealer': 0, 'Individual': 1}}, inplace = True)
df.replace({'Transmission': {'Manual': 0, 'Automatic': 1}}, inplace = True)
```

```python
# Input Feature - X
# Target feature - Y
X = df.drop(['Car_Name','Selling_Price'], axis = 1)
Y = df['Selling_Price']
```

```python
X
```

```python
Y
```

## 4. Train-test split

```python
# Splitting the data into Training and Testing Data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

## 5. Model Trainig

```python
# Model Training
model = LinearRegression()
```

```python
# Training the ML model with training data
model.fit(X_train, Y_train)
```

## 6. Model Evaluation

```python
# Prediction of Training data
training_data_prediction = model.predict(X_train)
# Finding accracy score of training data
training_data_accuracy = metrics.r2_score(Y_train, training_data_prediction)
print("Accuracy score of training data is", training_data_accuracy)
```

```python
# Prediction of Testing data
testing_data_prediction = model.predict(X_test)
# Finding accracy score of testing data
testing_data_accuracy = metrics.r2_score(Y_test, testing_data_prediction)
print("Accuracy score of testing data is", testing_data_accuracy)
```

```python
# Builiding a predictive system
input_data = (2011,0.826,6000,0,1,0,0)
input_data_array = np.asarray(input_data)
# Reshape to one data point
input_data_reshape = input_data_array.reshape(1,-1)
```

## 7. Prediction

```python
# Prediction
prediction = model.predict(input_data_reshape)
print('Selling price of the vehicle is', prediction)
```
